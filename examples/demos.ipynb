# tic toc
Mimics Matlab's tic toc in case you miss it

```python
from jupytertweak import tic, toc

tic()
print("Hello World...")
x = [i for i in range(10000)]
toc()
```

In Jupyter the easier way would be

```python
%%time
print("Hello World...")
x = [i for i in range(10000)]
```
